## 형태소 분석기 공통 메소드
- `morphs(string)` : 형태소 단위로 토큰화(tokenize)
    - stem : stemming 처리 여부 (기본: False)
- `pos(string)`: 품사 부착
    - stem : stemming 처리 여부 (기본: False)
- `nouns(string)` : 명사 추출하여 토큰화(tokenize)
- `tagset`: 형태소 분석기가 사용하는 품사태그 설명하는 속성.

# KoNLPy 형태소 분석기와 NLTK를 이용한 문서의 분석

```python
from nltk import Text
from konlpy.corpus import kolaw
from konlpy.tag import Okt
```

```python
law_txt = kolaw.open('constitution.txt').read()
```

```python
# 명사만 추출
okt = Okt()
noun_list = okt.nouns(law_txt)
len(noun_list)
```

```python
noun_list[:10]
```

```python
# 2글자 이상 명사들만 조회
noun_list = [noun for noun in noun_list if len(noun) > 1]
len(noun_list)
```

```python
# nltk.Text를 이용해 분석
text = Text(noun_list, '헌법')
text
```

```python
text.count("대한민국"), text.count('국민')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,7))
text.plot(15)
plt.show()
```

```python
plt.figure(figsize=(10,10))
text.dispersion_plot(['법률', '대통령','국가'])
plt.show()
```

```python
import nltk
noun_freq = text.vocab()
noun_freq = nltk.FreqDist(noun_list)
noun_freq
```

```python
print('총 토큰수:', noun_freq.N())
print('최빈도 토큰:', noun_freq.max())
print('단어 출연 횟수:', noun_freq.get('법률'))
print('단어 출연 비율:', noun_freq.freq('법률'))
```

```python
noun_freq.most_common()
```

```python
with open('news_text.txt', encoding='utf-8') as f:
    news_txt = f.read()
```

```python
news_txt
```

```python
from konlpy.tag import Okt
from nltk import Text, FreqDist
```

```python
okt = Okt()
news_token = okt.morphs(news_txt, stem=True) # 형태소 단위로 토큰화
news_token[:20]
```

```python
noun_list = okt.nouns(news_txt)
noun_list[:10]
```

```python
text = Text(noun_list, "뉴스기사")
text
```

```python
# 빈도수 선그래프
import matplotlib.pyplot as plt
plt.figure(figsize=(7,7))
```

```python
text.plot(10)
```

```python
plt.figure(figsize=(10, 10))
text.dispersion_plot(['인공위성', '관측', '천체'])
```

```python
fd = text.vocab()
fd = FreqDist(noun_list)
```

```python
fd.N() #전체토큰수
```

```python
fd.get('관측'), fd.get('천체'), fd['인공위성'] # 특정 토큰의 개수
```

```python
fd.freq('관측') #비율
```

```python
fd.max() #최빈도 토큰
```

```python
fd.most_common() # 빈도수 순서로 정렬
```

```python
import pandas as pd
df = pd.DataFrame(fd.most_common(), columns=['단어', '빈도수'])
df
```
